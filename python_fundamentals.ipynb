```python
# do first in terminal: $ conda install -c anaconda pillow

import urllib.request
import io
import PIL.Image

URL = 'http://www.gmit.ie/sites/all/themes/gmitpublic/images/gmit_logo.png'

with urllib.request.urlopen(URL) as url:
    imagefile = io.BytesIO(url.read())
    image = PIL.Image.open(imagefile)
    imagebw = image.convert('L')
    
display(image, imagebw)
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as pl

x1 = np.linspace(0, 4 * np.pi, 1000)
y1 = np.sin(x1)
y2 = np.cos(x1)

x2 = np.linspace(0, 4 * np.pi, 9)
y3 = np.sin(x2)
y3 = np.cos(x2)

pl.plot(x1, y1)
pl.plot(x2, y3, 'r.')
pl.plot(x1, y2, 'g')
pl.plot(x2, y3, 'k.')

pl.show()
```
